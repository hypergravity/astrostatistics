## 2. Matplotlib

````python
import matplotlib.pyplot as plt
````

````python
# line graph
x, y = np.arange(100), np.random.randn(100,)
plt.plot(x, y)
````

````python
# histogram
plt.hist(y);
````

````python
# bar graph
plt.bar(x, y)
````

````python
# scatter plot
plt.scatter(x, y, s=x, c=y)
````

````python
# contour
x, y = np.random.randn(1000,), np.random.randn(1000,)
counts, xedges, yedges, Image = plt.hist2d(x, y, bins=(20, 20))
plt.contour(counts, extent=(xedges[0], xedges[-1], yedges[0], yedges[-1]), colors='w')
````
