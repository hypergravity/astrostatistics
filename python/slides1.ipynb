```python
# Basic python
print("Hello world!")

print(type("Hello world!"))

print(type("Hello world!")==str)

# dir()
```

```python
# Built-in data types

# int
print(type(1))

# float
print(type(1.0))

# str
print(type("Hello World"))

# bool
print(type(False))
```

```python
# Built-in data types: list

alist = [1, 2.0, "3"]
print(alist)
print(type(alist))
print(len(alist))

alist.append(4)
print(alist)
```

```python
# how to access elements in a list

print(alist)
print(alist[1])
print(alist[::3])
print(alist[1:])
print(alist[:-1])
```

```python
# Built-in data types: tuple

atuple = (1, 2.0, "3")
print(atuple)
print(type(atuple))

# how to accesse elements in a list
print(atuple[1])
print(atuple[::2])
print(atuple[1:])
print(atuple[:-1])
```

```python
# Set & dictionary

aset = {1, 2.0, '3'}
print(aset)

adict = {1:"1.000", 2.0:"2.000", "3":"3.000"}
print(adict)
print(adict[1])
print(adict.keys(), adict.values())
```

# Control flow commands

```python
# for

for i in range(10):
    print(i)
```

```python
# for

print(alist)

for _ in alist:
    print(_)
```

```python
# for

print(adict)

for k, v in adict.items():
    print(k, v)
```

```python
# if elif else

data = [-2, 0, 2]

for _ in data:
    if _ < 0:
        print(_)
    elif _ == 0:
        print("exactly 0")
    else:
        print(_)
```

```python
# while

a = 10
while(a>0):
    print(a) 
    a-=3    
```

# numpy

```python
# how to import packages/modules

# 1
import numpy
print(numpy.sqrt(2))

# 2
import numpy as np
print(np.sqrt(2))

# 3
from numpy import sqrt
print(sqrt(2))
```

```python
# numpy.ndarray VS list

x_list = [1,2,3]

import numpy as np
x_array = np.array(x_list)

print(x_list, x_list*2)
print(x_array, x_array*2)
```

```python
# numpy

import numpy as np
a = np.arange(15).reshape(3, 5)
print("a = ", a)

print("the shape of a is ", a.shape)

print(a.ndim)

print(a.dtype.name)

print(a.itemsize)

print(a.size)

print(type(a))
```
